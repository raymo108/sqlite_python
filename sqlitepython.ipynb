```python
#https://www.youtube.com/watch?v=aG9LG_B5bSs

import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('train.csv')
conn = sqlite3.connect('phone.db')

c = conn.cursor()
c.execute('CREATE TABLE IF NOT EXISTS phonedata ({})'.format(','.join(df.columns)))
```

```python
for row in df.iterrows():
    sql = 'INSERT INTO phonedata({}) VALUES ({})'.format(','.join(df.columns), ','.join(['?']*len(df.columns)))
    c.execute(sql, tuple(row[1]))
conn.commit()
```
